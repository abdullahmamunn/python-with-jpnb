```python
from scipy.stats import norm
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [1, 2, 2, 3])
```

```python
# Draw a line graph
x = np.arange(-3, 3, 0.001)
plt.plot(x, norm.pdf(x))
plt.grid(True)
plt.show()
```

```python
# Multiple plots on one graph
x = np.arange(-3, 3, 0.001)
plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x, 1.0, 0.5))
plt.plot(x, norm.pdf(x, 2.0, 0.5))
plt.plot(x, norm.pdf(x, 3.0, 0.5))
plt.savefig('multiplot.png', dpi=300, bbox_inches='tight')
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Bangladesh', 'India', 'Sri Lanka', 'Pakistan'
sizes = [15, 30, 45, 10]
explode = (0.1, 0, 0, 0)  # only "explode" the 2nd slice (i.e. 'Bangladesh')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
labels = ['HTML-CSS', 'PHP', 'JAVA', 'C++', 'Python']
values = [40, 30, 25, 20, 35]
explode = (0, 0, 0.1, 0, 0)
colors = ['r','g','b','c','m']
plt.pie(values, labels = labels, colors = colors, explode = explode, autopct='%1.1f%%', shadow=True, startangle=40)
plt.title('My Programming Skills')
plt.savefig('pie-chart.png', dpi=300, bbox_inches='tight')
```

```python
# Bar chart
labels = ['HTML-CSS', 'PHP', 'JAVA', 'C++', 'Python']
values = [40, 30, 25, 20, 35]
explode = (0, 0, 0, 0, 0)
colors = ['r','g','b','c','m']
plt.bar(range(0,5), values, color = colors, label = labels)
# plt.pie(values, labels = labels, colors = colors, explode = explode, autopct='%1.1f%%', shadow=True, startangle=40)
plt.title('My Programming Skills')

plt.savefig('barchart.png', dpi=300, bbox_inches='tight')
```
